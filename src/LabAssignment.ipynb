1. Separate the variables that can be used and the ones that cannot be used.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler

data = '../data/wbc.csv' # path dataset
df = pd.read_csv(data) # load dataset
df.head() #Show the 5 first data by using df.head()
```

```python
df.info()
```

```python
unused_columns = ['id']
unnamed_columns = ['Unnamed: 32']
df.drop(unused_columns, axis=1, inplace=True)
df.drop(unnamed_columns, axis=1, inplace=True)
df.head()
```

2. Perform encoding on the "diagnosis" column.

```python
df = df[['diagnosis']]
df.head()
```

```python
le = LabelEncoder()
#encoding process
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df.head()
```

3. Standardize all columns with numeric values

```python
from sklearn.preprocessing import StandardScaler
data = '../data/wbc.csv' # path dataset
df = pd.read_csv(data) # load dataset

numeric_cols = df.select_dtypes(include=['number'])
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_cols)
# Replace the original data with standardized data
df[numeric_cols.columns] = scaled_data
```

```python
df.head()
```

4. Perform a stratified split of the data to create training and test data with an 80:20 ratio

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
# Split data into training (80%) and testing (20%)
df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Check the amount of data in each subset
print(f'Number of Original Data     :\n{df.diagnosis.value_counts()}')
print(f'Number of Training Data     :\n{df_train.diagnosis.value_counts()}')
print(f'Number of Testing Data      :\n{df_test.diagnosis.value_counts()}')
```
